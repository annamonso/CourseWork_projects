```python
import pandas as pd
import numpy as np 
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Load data

communities = pd.read_csv("../Input Files/communities.csv")["Communities"].tolist()
edges = pd.read_csv("../Input Files/edges.csv", usecols=["From", "To"])
print(edges.head())

```

```python
# Create an empty graph 
G = nx.Graph()

G.add_nodes_from(communities)
edges_list = [tuple(x) for x in edges.to_numpy()]
G.add_edges_from(edges_list)
# nx.draw(G)
nx.draw(G, with_labels=True, node_size=100, font_size=8)
plt.show()
```

```python
# Degree distribution
# Compute degree counts
degrees = [deg for _, deg in G.degree()]
degree_counts = np.bincount(degrees)
degrees_unique = np.arange(len(degree_counts))
fraction = degree_counts / sum(degree_counts)

# plot 
plt.hist(degrees, bins=range(1, max(degrees)+2), align='left', rwidth=0.8)
plt.xlabel('Degree')
plt.ylabel('Number of Communites')
plt.title('Degree Distribution')
plt.show()
```

```python
# Get nodes with degree 1
nodes_degree_1 = [node for node, deg in G.degree() if deg == 1]

# Get nodes with degree 8
nodes_degree_8 = [node for node, deg in G.degree() if deg == 8]

print("Communities with degree 1:", nodes_degree_1)
print("Communities with degree 8:", nodes_degree_8)
```
